```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load cleaned data
data = pd.read_csv('../data/cleaned_customer_data.csv')

# Pairplot to see relationships
sns.pairplot(data)
plt.show()

# Correlation matrix
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# Load clustered data
data = pd.read_csv('../data/clustered_customer_data.csv')

# Perform TSNE
tsne = TSNE(n_components=2, random_state=42)
tsne_results = tsne.fit_transform(data.select_dtypes(include=[np.number]))

data['TSNE-1'] = tsne_results[:,0]
data['TSNE-2'] = tsne_results[:,1]

# Visualize TSNE results
sns.scatterplot(x='TSNE-1', y='TSNE-2', hue='Cluster', data=data, palette='viridis')
plt.title('TSNE Plot of Customer Segments')
plt.show()
```
